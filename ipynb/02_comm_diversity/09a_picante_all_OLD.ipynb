# Goals

* Picante-based analyses on the dataset

From the [Picante tutorial](https://cran.r-project.org/web/packages/picante/vignettes/picante-intro.pdf):

> Positive SES values (mpd.obs.z > 0) and high quantiles (mpd.obs.p > 0.95)
indicate phylogenetic evenness, or a greater phylogenetic distance among co-occurring
species than expected. Negative SES values and low quantiles (mpd.obs.p < 0.05)
indicate phylogenetic clustering, or small phylogenetic distances among co-occurring
species than expected. MPD is generally thought to be more sensitive to tree-wide
patterns of phylogenetic clustering and eveness, while MNTD is more sensitive to
patterns of evenness and clustering closer to the tips of the phylogeny

# Load

```r
# sample metadata
meta = Fread(metadata_file) %>%
    mutate.(Sample = gsub('-', '.', Sample)) %>%
    filter.(number_reads >= 5e5)
meta
```

```r
# Bracken count table
brk = qs::qread(brk_file, nthreads=threads) %>%
    mutate.(name = gsub(' ', '_', name))
brk 
```

```r
# trees
## genome phylogeny
gtdb_phy = read.tree(phy_file)
gtdb_phy
## trait dendrogram
trt_phy = read.tree(phy_trt_file)
trt_phy
```

```r
trees = list('gen_phy' = phy_file, 
         'trt_bray' = phy_trt_file, 
         'cog' = phy_cog_file, 
         'cog_cat' = phy_cogcat_file, 
         'pfam' = phy_pfam_file, 
         'pfam_cat' = phy_pfamcat_file)
trees = trees %>%
    lapply(read.tree)
trees %>% names
```

## Checks

```r
# overlap of tip labels among trees
overlap(trees[[1]]$tip.label, trees[[2]]$tip.label)
```

```r
# overlap with tips 
overlap(trees[[1]]$tip.label, brk$name)
```

### Filtering count table

```r
brk %>% unique_n('Samples', Sample)
x = meta %>% pull.(Sample) %>% unique
brk = brk %>%
    filter.(Sample %in% x)

brk %>% unique_n('Samples', Sample)
brk
```

### Converting to a wide matrix

```r
m = brk %>%
    select.(Sample, name, Abundance) %>%
    pivot_wider.(name, Abundance) %>%
    mutate_across.(is.numeric, 
                   function(x) ifelse(is.na(x), 0, x)) %>%
    as.data.frame
rownames(m) = m$Sample
m$Sample = NULL
m
```

```r
# # -- debug --
# m = m[1:100,1:1000]
# m
```

### Filtering trees

```r
filter_tree = function(tree, m){
    to_rm = setdiff(tree$tip.label, colnames(m))
    tree = drop.tip(tree, to_rm)
    return(tree)
}
trees_f = trees %>%
    lapply(filter_tree, m=m)
trees_f
```

```r
# tree to dist matrix
trees_dist = trees_f %>%
    lapply(cophenetic)
trees_dist %>% names
```

# MPD

```r
# ses.mpd
## WARNING: SLOW
ses.mpd.func = function(tree_dist, m){
    ses.mpd(m, tree_dist, null.model='taxa.labels', abundance.weighted=FALSE, runs=99)
}

ses_mpd_res = Q(ses.mpd.func, tree_dist=trees_dist, const=list('m'=m), n_jobs=threads, verbose=FALSE)
ses_mpd_res %>% names
```

## Summary

```r
# to data.frame
lmer_res_f = lmer_res %>%
    lapply(function(x) tibble::rownames_to_column(x, 'variable')) %>%
    data.table::rbindlist(use.names=TRUE, idcol='hypothesis') %>%
    separate(hypothesis, c('dataset', 'metric', 'tree'), sep='_') %>%
    filter(dataset != 'w-g') %>%
    mutate(dataset = case_when(dataset == 'full' ~ '~ westernized +\n(1 | dataset)',
                               dataset == 'w-g-a' ~ '~ westernized +\nage + gender (1 | dataset)',
                               dataset == 'w-g-a-b' ~ '~ westernized +\nage+ gender + BMI (1 | dataset)'))
lmer_res_f
```

```r
p = lmer_res_f %>%
    mutate(P = formatC(`Pr(>F)`, format = "e", digits = 0)) %>%
    ggplot(aes(variable, `F value`, fill=tree)) +
    geom_bar(stat='identity', position='dodge') +
    geom_text(aes(y = `F value` + 2, label=P), 
              position = position_dodge(width = 1),
              size=3) +
    facet_grid(metric ~ dataset, scales='free_x', space='free_x') +
    labs(x='Coefficient') +
    theme_bw() 

p.dims(9,3.7)
plot(p)
```

  * Less functionally-constrained with increasing age
